```python
-- 코드를 입력하세요
SELECT CAR_ID, ROUND(AVG(DATEDIFF(END_DATE,START_DATE)+1),1) AS AVERAGE_DURATION
FROM CAR_RENTAL_COMPANY_RENTAL_HISTORY
GROUP BY CAR_ID
HAVING AVG(DATEDIFF(END_DATE,START_DATE)+1) >= 7
ORDER BY AVERAGE_DURATION DESC, CAR_ID DESC;

//DATEDIFF를 사용했을 경우 시작일과 마지막날짜의 차이만 계산하기 떄문에 +1을 진행해 시작일을 포함해줘야함
```

```python
-- 코드를 입력하세요
SELECT 
CASE 
WHEN PRODUCT_ID IN (1,2) THEN 'A1' 
WHEN PRODUCT_ID IN (3,4,5) THEN 'C3' 
WHEN PRODUCT_ID IN (6) THEN 'K1'
END AS CATEGORY,COUNT(*) AS PRODUCTS
FROM PRODUCT
GROUP BY CATEGORY
ORDER BY CATEGORY ASC;
-------------------------------------------------------------------
SELECT LEFT(PRODUCT_CODE, 2) AS CATEGORY,
       COUNT(*) AS PRODUCTS
FROM PRODUCT
GROUP BY LEFT(PRODUCT_CODE, 2)
ORDER BY CATEGORY ASC;

//처음에는 PRODUCT_ID의 첫 2글자만 추출해서 숫자에 맞게 A1,C3,K1을 설정해주면 될 줄 알아서 CASE문을 사용했지만 예제를 실행했을 때 6이상의 숫자들이 나와서 NULL값이 나와서
틀리기 되는 상황이 나와버렸다. 이 문제를 해결하기 위해서는 CASE문을 사용하지말고 앞 두글자를 추출해서 하면 되는 후자방법을 사용하면 된다.
```

```python
-- 코드를 입력하세요
SELECT CASE
WHEN PRICE BETWEEN 0 AND 9999 THEN '0'
WHEN PRICE BETWEEN 10000 AND 19999 THEN '10000'
WHEN PRICE BETWEEN 20000 AND 29999 THEN '20000'
WHEN PRICE BETWEEN 30000 AND 39999 THEN '30000'
END AS PRICE_GROUP,
COUNT(*) AS PRODUCTS
FROM PRODUCT 
GROUP  BY PRICE_GROUP
ORDER BY PRICE_GROUP ASC;
-----------------------------------------------------

SELECT 
    (PRICE DIV 10000) * 10000 AS PRICE_GROUP,
    COUNT(*) AS PRODUCTS
FROM PRODUCT
GROUP BY PRICE_GROUP
ORDER BY PRICE_GROUP ASC;

//전자처럼 해버리면 출력시 해당 조건값만 있는게 아닐 경우에는 조건값이 아닌경우들이 카운트가 되버려서 실행시 표시가 안되는 모습이 보여지기 때문에, 후자와 같이 가격을 10000으로
나누어 한자리수의 정수로 만든 후에 10000을 곱해 구간을 정해준다. 
```

```python
-- 코드를 작성해주세요
SELECT CASE
WHEN MONTH(DIFFERENTIATION_DATE) BETWEEN 1 AND 3 THEN '1Q'
WHEN MONTH(DIFFERENTIATION_DATE) BETWEEN 4 AND 6 THEN '2Q'
WHEN MONTH(DIFFERENTIATION_DATE) BETWEEN 7 AND 9 THEN '3Q'
WHEN MONTH(DIFFERENTIATION_DATE) BETWEEN 10 AND 12 THEN '4Q'
END AS QUARTER,
COUNT(*) AS ECOLI_COUNT
FROM ECOLI_DATA 
GROUP BY QUARTER
ORDER BY QUARTER ASC;
-------------------------------------------------------------------
SELECT CONCAT(QUARTER(DIFFERENTIATION_DATE),'Q') AS QUARTER,COUNT(*) AS ECOLI_COUNT
FROM ECOLI_DATA
GROUP BY QUARTER
ORDER BY QUARTER;

//두개 방식 둘 다 가능하다. 
```

```python
-- 코드를 작성해주세요
SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR , 
MAX(SIZE_OF_COLONY) OVER (PARTITION BY YEAR(DIFFERENTIATION_DATE)) - SIZE_OF_COLONY AS YEAR_DEV
,ID
FROM ECOLI_DATA 
ORDER BY YEAR ASC, YEAR_DEV ASC;

//OVER (PARTITION BY YEAR(DIFFERENTIATION_DATE)) 를 통해서 각 년도별로 그룹을 나누게 되는데 MAX(SIZE_OF_COLONY) 연도안에서 가장 큰값을 뽑아내게 된다. 이후 자기자신을
빼서 편차를 구할 수 있게 된다.

SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR , (M.MAX_SIZE - E.SIZE_OF_COLONY) AS YEAR_DEV, E.ID
FROM ECOLI_DATA E
JOIN 
    (SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR ,MAX(SIZE_OF_COLONY) AS MAX_SIZE
     FROM ECOLI_DATA
     GROUP BY YEAR(DIFFERENTIATION_DATE)
    ) M ON YEAR(E.DIFFERENTIATION_DATE) = M.YEAR
ORDER BY YEAR ASC, YEAR_DEV ASC;

//서브쿼리 사용해서 해결
```

```python
SELECT C.ID, C.GENOTYPE, P.GENOTYPE AS PARENT_GENOTYPE
FROM ECOLI_DATA C JOIN ECOLI_DATA P ON C.PARENT_ID = P.ID
WHERE C.GENOTYPE & P.GENOTYPE = P.GENOTYPE
ORDER BY C.ID ASC;

//
C.GENOTYPE & P.GENOTYPE = P.GENOTYPE #자식이 부모의 형질을 가지는 것만 출력해야하기 때문에 해당 조건으로 진행
ECOLI_DATA라는 테이블안에 부모와 자식의 데이터가 있기 떄문에 같은 테이블을 두 번 불러서 자식(C) ↔ 부모(P) 관계를 연결 해야 해서 SELF조인을 진행
ECOLI_DATA P ON C.PARENT_ID = P.ID 자식이 부모의 ID값을 부모는 자신의 ID값이 일치해야 연결이된다.
```
