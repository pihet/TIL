```python
-- 코드를 작성해주세요
SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR , 
MAX(SIZE_OF_COLONY) OVER (PARTITION BY YEAR(DIFFERENTIATION_DATE)) - SIZE_OF_COLONY AS YEAR_DEV
,ID
FROM ECOLI_DATA 
ORDER BY YEAR ASC, YEAR_DEV ASC;

//OVER (PARTITION BY YEAR(DIFFERENTIATION_DATE)) 를 통해서 각 년도별로 그룹을 나누게 되는데 MAX(SIZE_OF_COLONY) 연도안에서 가장 큰값을 뽑아내게 된다. 이후 자기자신을
빼서 편차를 구할 수 있게 된다.

SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR , (M.MAX_SIZE - E.SIZE_OF_COLONY) AS YEAR_DEV, E.ID
FROM ECOLI_DATA E
JOIN 
    (SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR ,MAX(SIZE_OF_COLONY) AS MAX_SIZE
     FROM ECOLI_DATA
     GROUP BY YEAR(DIFFERENTIATION_DATE)
    ) M ON YEAR(E.DIFFERENTIATION_DATE) = M.YEAR
ORDER BY YEAR ASC, YEAR_DEV ASC;

//서브쿼리 사용해서 해결
```

```python
-문제-
보호소에서는 몇 시에 입양이 가장 활발하게 일어나는지 알아보려 합니다. 0시부터 23시까지, 
각 시간대별로 입양이 몇 건이나 발생했는지 조회하는 SQL문을 작성해주세요. 이때 결과는 시간대 순으로 정렬해야 합니다.

#해당문제를 풀었을 때 발생하는 문제점은 테스트값에 0시부터 23시중 없는 시간대가 발생했기 때문에 예시문과 동일하게 나오지 않고, 시간대가 드문드문 있었다는 점
이 문제점을 해결하기 위해 SET을 통한 테이블에 없는 값을 강제로 만들거나 WITH RECURSIVE를 통해 LEFT JOIN을 하는 방법으로 해결하는 것.

SET @hour := -1;
SELECT (@hour := @hour + 1) AS HOUR, (SELECT COUNT(@hour) FROM ANIMAL_OUTS WHERE HOUR(DATETIME) = @hour)AS COUNT
FROM ANIMAL_OUTS 
WHERE @hour < 23;
//SET을 사용했을 때 @변수명 :=(SQL에서는 =을 :=으로 정의해야함) 

WITH RECURSIVE hours AS (
    SELECT 0 AS h
    UNION ALL
    SELECT h + 1
    FROM hours
    WHERE h < 23
)
SELECT h, COUNT(ANIMAL_ID) AS COUNT
FROM hours H LEFT JOIN ANIMAL_OUTS  A ON H.h = HOUR(A.DATETIME)
GROUP BY h
ORDER BY h;
```

```python
SELECT 
    H.HISTORY_ID,
    ROUND(
        C.DAILY_FEE * (DATEDIFF(H.END_DATE, H.START_DATE) + 1) * (100 - IFNULL(P.DISCOUNT_RATE, 0)) / 100
    ) AS FEE
FROM 
    CAR_RENTAL_COMPANY_RENTAL_HISTORY H
JOIN 
    CAR_RENTAL_COMPANY_CAR C ON H.CAR_ID = C.CAR_ID
LEFT JOIN 
    CAR_RENTAL_COMPANY_DISCOUNT_PLAN P 
    ON C.CAR_TYPE = P.CAR_TYPE 
    AND P.DURATION_TYPE = (
        CASE 
            WHEN DATEDIFF(H.END_DATE, H.START_DATE) + 1 >= 90 THEN '90일 이상'
            WHEN DATEDIFF(H.END_DATE, H.START_DATE) + 1 >= 30 THEN '30일 이상'
            WHEN DATEDIFF(H.END_DATE, H.START_DATE) + 1 >= 7  THEN '7일 이상'
            ELSE NULL 
        END
    )
WHERE 
    C.CAR_TYPE = '트럭'
ORDER BY 
    FEE DESC, 
    H.HISTORY_ID DESC;

#FROM절에서 DURATION_TYPE으로 INNER JOIN을 하지 않는 이유는 할인받을 수 있는 종류가 3개이기 때문에 1개의 트럭값을 찾을려고 했을 때 3개의 트럭할인이 붙기떄문에 데이터
가 증가되어서 뻥튀기 되는 현상이 나와버린다. 그래서 FROM절에서 CASE문을 통해 조건을 건 후 SELECT에서 FEE에 대한 값을 지정하는데, 7일 미만인 경우에는
IFNULL을 통해 필터링이 돼 100%값으로 진행된다. 왜 그렇냐면 CASE문에서 7일 미만은 설정하지 않았기 때문에, 조인을 하게 되면 NULL값이 된다. IFNULL은 해당값이 NULL이면 
뒤에 값으로 반환하기 때문에 NULL인 순간 0을 반환해줘서 100%로 비용이 됨
```
