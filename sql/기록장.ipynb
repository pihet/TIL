```python
-- 코드를 작성해주세요
SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR , 
MAX(SIZE_OF_COLONY) OVER (PARTITION BY YEAR(DIFFERENTIATION_DATE)) - SIZE_OF_COLONY AS YEAR_DEV
,ID
FROM ECOLI_DATA 
ORDER BY YEAR ASC, YEAR_DEV ASC;

//OVER (PARTITION BY YEAR(DIFFERENTIATION_DATE)) 를 통해서 각 년도별로 그룹을 나누게 되는데 MAX(SIZE_OF_COLONY) 연도안에서 가장 큰값을 뽑아내게 된다. 이후 자기자신을
빼서 편차를 구할 수 있게 된다.

SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR , (M.MAX_SIZE - E.SIZE_OF_COLONY) AS YEAR_DEV, E.ID
FROM ECOLI_DATA E
JOIN 
    (SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR ,MAX(SIZE_OF_COLONY) AS MAX_SIZE
     FROM ECOLI_DATA
     GROUP BY YEAR(DIFFERENTIATION_DATE)
    ) M ON YEAR(E.DIFFERENTIATION_DATE) = M.YEAR
ORDER BY YEAR ASC, YEAR_DEV ASC;

//서브쿼리 사용해서 해결
```

```python
SELECT A.ID, A.NAME, A.HOST_ID
FROM PLACES A JOIN (SELECT HOST_ID AS H
FROM PLACES 
GROUP BY HOST_ID
HAVING COUNT(HOST_ID) >= 2) B ON A.HOST_ID = B.H
ORDER BY A.ID;
//FROM절에서 서브쿼리를 통한 필터링 방식인데, 필터링을 거쳤기 떄문에 A테이블에서 조건에 맞지 않는 행들은 전부 지워져버린다. 그래서 A.HOST_ID를 해도되는거였고, B.H를 해도 
A.HOST_ID와 동일한 결과값이 나오지만 A.HOST_ID를 하는게 관습적이라고 함.
--------------------------------------------------------------------
SELECT ID, NAME, HOST_ID
FROM PLACES
WHERE HOST_ID IN (
    SELECT HOST_ID
    FROM PLACES
    GROUP BY HOST_ID
    HAVING COUNT(ID) >= 2
)
ORDER BY ID;
//이건 WHERE절에서 조건을 걸어서 필터링을 하는 방식 
```

```python
SELECT COUNT(*) AS FISH_COUNT,MAX(LENGTH) AS MAX_LENGTH,FISH_TYPE
FROM FISH_INFO
GROUP BY FISH_TYPE
HAVING AVG(CASE WHEN LENGTH <= 10 THEN 10 
           ELSE LENGTH
           END) >= 33
ORDER BY FISH_TYPE ASC;

//해당문제에선 길이가 10미만인 경우에는 NULL로 취급하기 때문에 HAVING절에서 길이가 10미만인걸 10으로 취급하는게 불가능하기 떄문에 위와 같은 조건을 걸면 부분적으로 틀리는 
경우가 발생하게 된다. 이를 해결하기 위해서는 IFNULL을 사용해서 해결 할 수 있었다.
-- 코드를 작성해주세요
SELECT COUNT(*) AS FISH_COUNT,MAX(LENGTH) AS MAX_LENGTH,FISH_TYPE
FROM FISH_INFO
GROUP BY FISH_TYPE
HAVING AVG(IFNULL(LENGTH,10)) >= 33
ORDER BY FISH_TYPE ASC;

#CASE문을 사용한다고 했을 땐
CASE WHEN LENGTH IS NULL TEHN 10
     WHEN LENGTH <= 10 THEN 10
     ELSE LENGTH
     END
```

```python
-- 코드를 작성해주세요
SELECT ID, 
CASE NTILE(4) OVER (ORDER BY SIZE_OF_COLONY DESC)
WHEN 1 THEN 'CRITICAL'
WHEN 2 THEN 'HIGH'
WHEN 3 THEN 'MEDIUM'
WHEN 4 THEN 'LOW'
END AS COLONY_NAME
FROM ECOLI_DATA
ORDER BY ID ASC;
//크기를 기준으로 할 땐 QUAETER를 사용하는게 아니라 NTILE을 이용해서 구별해야한다. QUAETER함수는 날짜에서 분기를 뽑는 함수이기 때문에 NTILE사용해야한다.
```
